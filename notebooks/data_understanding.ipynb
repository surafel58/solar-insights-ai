# Solar Insights AI: Data Understanding

This Jupyter notebook aims to explore and analyze solar energy data from the Benin-Malanville, Sierraleone-Bumbuna, and Togo-Dapaong_qc datasets. The objective is to gain insights into the dataset's features and basic information. 

Key steps in this analysis include:
1. Loading and inspecting the dataset
2. Examining basic statistics and data distributions
3. Checking for missing values


```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Benin dataset
df_benin = pd.read_csv('data/benin-malanville.csv')

# Load the Sierra Leone dataset
df_sierraleone = pd.read_csv('data/sierraleone-bumbuna.csv')

# Load the Togo dataset
df_togo = pd.read_csv('data/togo-dapaong_qc.csv')
```

```python
# Display the first five rows
print("First five rows of the dataset:")
display(df_benin.head())
```

```python
#remove comment column 
del df_benin['Comments']

# Get a summary of the data
print("\nDataFrame Info:")
df_benin.info()
```

```python
# Get summary statistics
print("\nSummary Statistics:")
display(df_benin.describe())
```

```python
# Check for missing values
print("\nMissing Values in Each Column:")
missing_values = df_benin.isnull().sum()
display(missing_values)
```

```python
# Visualize missing values
plt.figure(figsize=(12, 6))
sns.heatmap(df_benin.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

### Key Findings
1. **Data Completeness**:
   - The dataset is complete with no missing values in any of the columns.

2. **Data Distribution**:
   - The summary statistics provide insights into the distribution of the data. For example:
     - `GHI` has a mean of approximately 240.56 and a standard deviation of 331.13.
     - `DNI` has a mean of approximately 167.19 and a standard deviation of 261.71.
     - `Tamb` has a mean of approximately 28.18 and a standard deviation of 5.92 degree celsius.
     - `RH` has a mean of approximately 54.49% and a standard deviation of 28.07%.

3. **Data Range**:
   - The data ranges for various columns are provided, such as:
     - `GHI` ranges from -12.9 to 1413.0.
     - `DNI` ranges from -7.8 to 952.3.
     - `Tamb` ranges from 11.0 to 43.8 degree celsius.
     - `RH` ranges from 2.1% to 100.0%.
